```python
#Establishing the conditional for the High Risk classification

# Condition 1: If income is 3 times lower than credit amount

condition_1 = risk_tg.client_income < (risk_tg.credit_amount / 3)

# Condition 2: If external scoring lower than 0.5

condition_2 = (risk_tg['ext_score_1'] > 0.5) & (risk_tg['ext_score_2'] > 0.5) & (risk_tg['ext_score_3'] > 0.5)

# Condition 3: If number of enquiries are more than 4 by year

condition_3 = risk_tg.req_year > 4

# Condition 4: If client living in a region and city with rating 2

condition_4 = (risk_tg.region_rating == 2) & (risk_tg.city_rating == 2)

# Condition 5: If loan type is categorized as 'cash loan'

condition_5 = risk_tg.loan_type == 'Cash loans'

# Condition 6: If the client is the ownership or live with parent

condition_6 = (risk_tg.housing_situation == 'Ownership') | (risk_tg.housing_situation == 'Living with parents')

# Condition 7: If the client is currently working

condition_7 = risk_tg.client_income_type == 'Working'

# Condition 8: If the client is currently married

condition_8 = risk_tg.family_status == 'Married'

# Condition 9: If the client has low education level

condition_9 = risk_tg.education == 'Secondary'

# Condition 10: If the client has not children

condition_10 = risk_tg.children == 0
```

```python
# First test with 10 conditions

# Condiciones para High Risk
condition_1 = risk_tg.client_income < (risk_tg.credit_amount / 3)
condition_2 = (risk_tg['ext_score_1'] > 0.5) & (risk_tg['ext_score_2'] > 0.5) & (risk_tg['ext_score_3'] > 0.5)
condition_3 = risk_tg.req_year > 4
condition_4 = (risk_tg.region_rating == 2) & (risk_tg.city_rating == 2)
condition_5 = risk_tg.loan_type == 'Cash loans'
condition_6 = (risk_tg.housing_situation == 'Ownership') | (risk_tg.housing_situation == 'Living with parents')
condition_7 = risk_tg.client_income_type == 'Working'
condition_8 = risk_tg.family_status == 'Married'
condition_9 = risk_tg.education == 'Secondary'
condition_10 = risk_tg.children == 0

# Asignar valores basados en condiciones
risk_tg['risk_category'] = np.select(
    [
        condition_1,
        condition_2,
        condition_3,
        condition_4,
        condition_5,
        condition_6,
        condition_7,
        condition_8,
        condition_9,
        condition_10
    ],
    ['High Risk'] * 10,  # Valor asignado si la condición es verdadera
    default='Low Risk'   # Valor asignado si ninguna condición es verdadera
)
```

```python
# Results provide us a bad classification

risk_tg.risk_category.value_counts()
```

```python
# Second test with 8 conditions


# Condiciones para High Risk
condition_1 = risk_tg.client_income < (risk_tg.credit_amount / 3)
condition_2 = (risk_tg['ext_score_1'] > 0.6) & (risk_tg['ext_score_2'] > 0.6) & (risk_tg['ext_score_3'] > 0.6)
condition_3 = risk_tg.req_year > 4
condition_4 = (risk_tg.region_rating == 2) & (risk_tg.city_rating == 2)
condition_5 = risk_tg.loan_type == 'Cash loans'
condition_6 = (risk_tg.housing_situation == 'Ownership') | (risk_tg.housing_situation == 'Living with parents')
condition_7 = risk_tg.client_income_type == 'Working'
condition_8 = risk_tg.family_status == 'Married'


# Asignar valores basados en condiciones
risk_tg['risk_category_2'] = np.select(
    [
        condition_1,
        condition_2,
        condition_3,
        condition_4,
        condition_5,
        condition_6,
        condition_7,
        condition_8,
            ],
    ['High Risk'] * 8,  # Valor asignado si la condición es verdadera
    default='Low Risk'   # Valor asignado si ninguna condición es verdadera
)
```

```python
# Results provide us a bad classification

risk_tg.risk_category_2.value_counts()
```

```python
# Third test with 4 conditions

# Condiciones para High Risk
condition_1 = risk_tg.client_income < (risk_tg.credit_amount / 3)
condition_2 = (risk_tg['ext_score_1'] > 0.6) & (risk_tg['ext_score_2'] > 0.6) & (risk_tg['ext_score_3'] > 0.6)
condition_3 = risk_tg.req_year > 4
condition_4 = (risk_tg.region_rating == 2) & (risk_tg.city_rating == 2)

# Asignar valores basados en condiciones
risk_tg['risk_category_3'] = np.select(
    [
        condition_1,
        condition_2,
        condition_3,
        condition_4,
        
            ],
    ['High Risk'] * 4,  # Valor asignado si la condición es verdadera
    default='Low Risk'   # Valor asignado si ninguna condición es verdadera
)
```

```python

risk_tg['risk_category_3'].value_counts()
```

```python
# Fourth test with 3 conditions

condition_1 = risk_tg.client_income < (risk_tg.credit_amount / 3)
condition_2 = (risk_tg['ext_score_1'] > 0.6) & (risk_tg['ext_score_2'] > 0.6) & (risk_tg['ext_score_3'] > 0.6)
condition_3 = risk_tg.req_year > 4

# Asignar valores basados en condiciones
risk_tg['risk_category_4'] = np.select(
    [
        condition_1,
        condition_2,
        condition_3,
                    ],
    ['High Risk'] * 3,  # Valor asignado si la condición es verdadera
    default='Low Risk'   # Valor asignado si ninguna condición es verdadera
)
```

```python
risk_tg['risk_category_4'].value_counts()
```

```python
# Check the model with confusion matrix and classification Table

from sklearn.metrics import confusion_matrix, classification_report
```

```python
# Contingence Table

contingency_table = pd.crosstab(risk_tg['target'], risk_tg['risk_category_4'], margins=True)

print(contingency_table)

print()

# Confusion Matrix

conf_matrix = confusion_matrix(risk_tg['target'], risk_tg['risk_category_4'])

# Mostrar la matriz de confusión
print(conf_matrix)

print()

# Classification report
print(classification_report(risk_tg['target'], risk_tg['risk_category_4']))
```

```python
# Sixth test modifying numerical values on condition

condition_1 = risk_tg.client_income < (risk_tg.credit_amount / 3.5)
condition_2 = (risk_tg['ext_score_1'] > 0.7) & (risk_tg['ext_score_2'] > 0.7) & (risk_tg['ext_score_3'] > 0.7)
condition_3 = risk_tg.req_year > 5

risk_tg['risk_category_5'] = np.select(
    [
        condition_1,
        condition_2,
        condition_3,
                    ],
    ['High Risk'] * 3,  
    default='Low Risk'   
)
```
