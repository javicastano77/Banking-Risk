```python
# Paso 0: Importar las librerias necesarias para realizar el análisis del dataset limpio

import pandas as pd

import matplotlib.pyplot as plt

import seaborn as sns
```

```python
# Paso 0: Ajustar el display del Dataframe para poder desplegar el total de columnas del dataset original.

pd.options.display.max_columns = 40
pd.options.display.max_rows = 122
```

```python
# Paso 1: Crear un nuevo Dataframe del dataset limpio.

risk = pd.read_csv('risk_df.csv')

risk
```

```python
# Paso 2: Analizar el perfil tipo de la variable 'target', en base a variables sociodemográficas.

risk.groupby('target').agg({'gender':lambda x: x.mode().iloc[0],'children':'mean','education':lambda x: x.mode().iloc[0],'family_status':lambda x: x.mode().iloc[0],'client_ocupation':lambda x: x.mode().iloc[0],'client_income_type':lambda x: x.mode().iloc[0]}) 
```

```python
# Paso 3:  Analizar el perfil tipo de la variable 'target', en base a variables de compra.

risk.groupby('target').agg({'loan_type': lambda x:x.mode().iloc[0],'own_car':lambda x: x.mode().iloc[0],'own_realty':lambda x: x.mode().iloc[0],'housing_situation':lambda x: x.mode().iloc[0],'car_age':'mean'}).round(2)
```

```python
# Paso 4: Analizar el perfil del target, en base a 'client_income', 'credit_amount' y 'loan_annuity'. 

risk.groupby('target').agg({'client_income': ['mean','std'],'credit_amount':['mean','std'],'loan_annuity':['mean','std']}).round(2)
```

```python
# Paso 5: Analizar el perfil del target y su distribución en género, en base a 'client_income', 'credit_amount' y 'loan_annuity'.

risk.groupby(['target','gender',]).agg({'client_income': ['count','mean','std'],'credit_amount':['mean','std'],'loan_annuity':['mean','std']}).round(2)
```

```python
# Paso 6: Analizar el perfil del target y su distribución en educación, en base a 'client_income', 'credit_amount' y 'loan_annuity'.

risk.groupby(['target','education',]).agg({'client_income': ['count','mean','std','max'],'credit_amount':['mean','std'],'loan_annuity':['mean','std']}).round(2)
```

```python
# Paso 7: Analizar el perfil del target y su distribución de estado civil, en base a 'client_income', 'credit_amount' y 'loan_annuity'.

risk.groupby(['target','family_status',]).agg({'client_income': ['count','mean','std','max'],'credit_amount':['mean','std',],'loan_annuity':['mean','std',]}).round(2)
```

```python
# Paso 8: Analizar el perfil del target y el área de trabajo, en base a 'client_income', 'credit_amount' y 'loan_annuity'.

risk.groupby(['target','client_ocupation',]).agg({'client_income': ['count','mean','std',],'credit_amount':['mean','std',],'loan_annuity':['mean','std',]}).round(2)
```

```python
# Paso 9: Analizar el perfil del target y el tipo de ocupación, en base a 'client_income', 'credit_amount' y 'loan_annuity'.

risk.groupby(['target','client_income_type',]).agg({'client_income': ['count','mean','std',],'credit_amount':['mean','std',],'loan_annuity':['mean','std',]}).round(2)
```

```python
# Paso 10:  Analizar el perfil del target y el tipo de vivienda, en base a 'client_income', 'credit_amount' y 'loan_annuity'.

risk.groupby(['target','housing_situation',]).agg({'client_income': ['count','mean','std',],'credit_amount':['mean','std',],'loan_annuity':['mean','std',]}).round(2)
```

```python
# Paso 11:  Analizar el perfil del target y el tipo de préstamo, en base a 'client_income', 'credit_amount' y 'loan_annuity'.

risk.groupby(['target','loan_type',]).agg({'client_income': ['count','mean','std',],'credit_amount':['mean','std',],'loan_annuity':['mean','std',]}).round(2)
```

```python
# Paso 12: Analizar el target en base a los scoring externos.

risk.groupby('target').agg({'ext_score_1': ['min','max','mean','std',],'ext_score_2':['min','max','mean','std',],'ext_score_3':['min','max','mean','std',]}).T.round(2)
```

```python
# Paso 13: Analizar el target en base a las variables de observación (defaulter y observable).

risk.groupby('target').agg({'obs_30': ['mean','std','max'],'def_30':['mean','std','max'],'obs_60':['mean','std','max'],'def_60':['mean','std','max']}).T.round(2)
```

```python
# Paso 14.1: Analizar el target en base a las variables de consulta al Área de Riesgo.

risk.groupby('target').agg({'req_hour': ['mean','std'],'req_day':['mean','std'],'req_week':['mean','std'],'req_mon':['mean','std'],'req_qrt':['mean','std'],'req_year':['mean','std']}).T.round(2)
```

```python
# Paso 14.2: Analizar el target en base a las variables de consulta al Área de Riesgo.(min y max)

risk.groupby('target').agg({'req_hour': ['min','max'],'req_day':['min','max'],'req_week':['min','max'],'req_mon':['min','max'],'req_qrt':['min','max'],'req_year':['min','max']}).T.round(2)
```

```python
# Paso 15:  Analizar el perfil del target distribuido por el rating de region, en base a 'client_income', 'credit_amount' y 'loan_annuity'.

risk.groupby(['target','region_rating',]).agg({'client_income': ['count','mean','std',],'credit_amount':['mean','std',],'loan_annuity':['mean','std',]}).round(2)
```

```python
# Paso 16:  Analizar el perfil del target distribuido por el rating de ciudad, en base a 'client_income', 'credit_amount' y 'loan_annuity'.

risk.groupby(['target','city_rating',]).agg({'client_income': ['count','mean','std',],'credit_amount':['mean','std',],'loan_annuity':['mean','std',]}).round(2)
```
