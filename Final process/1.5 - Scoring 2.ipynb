```python
# Paso 0: Importar las librerias necesarias para realizar el análisis del modelo

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report
import numpy as np
```

```python
# Paso 0: Ajustar el display del Dataframe para poder desplegar el total de columnas del dataset original.

pd.options.display.max_columns = 70
pd.options.display.max_rows = 122
```

```python
# Paso 1: Crear un nuevo Dataframe del dataset limpio.

risk_tg = pd.read_csv('risk_df.csv')
```

```python
# Paso 2: Desplegar el Dataframe para chequear las variables iniciales

risk_tg
```

```python
# Paso 3: Crear nueva variable para analizar el crédito pendiente en años.

# Calculo: dividir el 'credit_amount' entre el 'loan_annuity'(amortización anual)

risk_tg['loan_year']= risk_tg.credit_amount / risk_tg.loan_annuity

# Redondear al valor alto el cálculo, dado que los decimales equivalen a un año más

risk_tg.loan_year = np.ceil(risk_tg.loan_year).astype(int)

# Posicionar la nueva variable en el DF

risk_tg.insert(10, 'loan_year', risk_tg.pop('loan_year'))
```

```python
# Paso 4: Crear nueva variable para analizar la amortización mensual.

# Calculo: dividir el 'loan_annuity' entre 12 meses

risk_tg['loan_mon']= (risk_tg.loan_annuity / 12).round(1)

# Posicionar la nueva variable en el DF

risk_tg.insert(10, 'loan_mon', risk_tg.pop('loan_mon'))
```

```python
# Paso 5: Identificar el mínimo y máximo de duración de los créditos del dataset.

print(risk_tg.loan_year.max())
print(risk_tg.loan_year.min())
```

```python
# Paso 6: Crear nueva variable para clasificar los tipos de clientes en base a la duración del préstamo.

# Establecer condiciones con las que igualar las categorías

condition_1 = risk_tg['loan_year'] < 10
condition_2 = (risk_tg['loan_year'] >= 10) & (risk_tg['loan_year'] < 20)
condition_3 = (risk_tg['loan_year'] >= 20) & (risk_tg['loan_year'] < 30)
condition_4 = (risk_tg['loan_year'] >= 30) & (risk_tg['loan_year'] < 40)
condition_5 = risk_tg['loan_year'] >= 40

# Creación de la variable en el Dataframe

risk_tg.loc[condition_1, 'loan_term'] = 'Flash'
risk_tg.loc[condition_2, 'loan_term'] = 'Low Term'
risk_tg.loc[condition_2, 'loan_term'] = 'Middle Term'
risk_tg.loc[condition_3, 'loan_term'] = 'Long Term'
risk_tg.loc[condition_4, 'loan_term'] = 'Methuselah'

# Posicionar la nueva variable en el DF

risk_tg.insert(12, 'loan_term', risk_tg.pop('loan_term'))
```

```python
# Paso 7: Crear nueva variable para la probabilidad de defaul inicial.

risk_tg['def_initial'] = 0.00  

# Al no poder calcular la probabilidad acumulada y asumiendo que conforme mayor duración, mayor riesgo, asignamos un valor inicial escalado de menos a mas en base a la duración:

risk_tg.loc[risk_tg['loan_term'] == 'Low Term', 'def_initial'] = 0.10
risk_tg.loc[risk_tg['loan_term'] == 'Middle Term', 'def_initial'] = 0.15
risk_tg.loc[risk_tg['loan_term'] == 'Long Term', 'def_initial'] = 0.20
risk_tg.loc[risk_tg['loan_term'] == 'Methuselah', 'def_initial'] = 0.25
risk_tg.loc[risk_tg['loan_term'] == 'Flash', 'def_initial'] = 0.05

# Redondear los valores que puedo coger esta variable:

risk_tg.def_initial = risk_tg.def_initial.round(2)

# Posicionar la nueva variable en el DF:

risk_tg.insert(13, 'def_initial', risk_tg.pop('def_initial'))
```

```python
# Paso 8: Crear una proyección de 5 a 25 años con probabilidades acumuladas para cada tipo de prestamo.

# Cáculo: para cada salto de año, añadimos un ratio del 0.10. Se crea una variable para cada salto

risk_tg['def_5'] = risk_tg['def_initial'] + 0.10
risk_tg['def_10'] = risk_tg['def_initial'] + 0.20
risk_tg['def_15'] = risk_tg['def_initial'] + 0.30
risk_tg['def_20'] = risk_tg['def_initial'] + 0.40

# Posicionar las nuevas variables en el DF:

risk_tg.insert(14, 'def_5', risk_tg.pop('def_5'))
risk_tg.insert(15, 'def_10', risk_tg.pop('def_10'))
risk_tg.insert(16, 'def_15', risk_tg.pop('def_15'))
risk_tg.insert(17, 'def_20', risk_tg.pop('def_20'))

# Desplegar el Dataframe con las nuevas variables

risk_tg
```

```python
# Paso 9: Establecer condicones para generar la nueva variable de target

risk_tg['category'] = 'Low Risk'
conditions = (risk_tg['def_initial'] >= 0.6) | (risk_tg['def_5'] >= 0.6) | (risk_tg['def_10'] >= 0.6) | (risk_tg['def_15'] >= 0.6) | (risk_tg['def_20'] >= 0.6)
risk_tg.loc[conditions, 'category'] = 'High Risk'
```

```python
risk_tg.to_csv('Model_2.csv', index=False)
```
