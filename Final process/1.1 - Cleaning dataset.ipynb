```python
# Paso 4: Tras el primer análisis, eliminar variables sociodemográficas que no son determinantes para nuestro análisis.


risk_1 = risk_df.drop((['DAYS_BIRTH','DAYS_EMPLOYED','NAME_TYPE_SUITE','REGION_POPULATION_RELATIVE', 
                        'DAYS_REGISTRATION','DAYS_ID_PUBLISH','FLAG_MOBIL','FLAG_EMP_PHONE','FLAG_WORK_PHONE',
                        'FLAG_CONT_MOBILE','FLAG_PHONE','WEEKDAY_APPR_PROCESS_START','HOUR_APPR_PROCESS_START', 
                        'REG_REGION_NOT_LIVE_REGION', 'REG_REGION_NOT_WORK_REGION', 'LIVE_REGION_NOT_WORK_REGION', 
                        'REG_CITY_NOT_LIVE_CITY', 'REG_CITY_NOT_WORK_CITY','LIVE_CITY_NOT_WORK_CITY', 'FLAG_EMAIL',
                        'APARTMENTS_AVG', 'BASEMENTAREA_AVG', 'YEARS_BEGINEXPLUATATION_AVG', 'YEARS_BUILD_AVG',
                        'COMMONAREA_AVG', 'ELEVATORS_AVG', 'ENTRANCES_AVG', 'FLOORSMAX_AVG', 'FLOORSMIN_AVG',
                        'LANDAREA_AVG', 'LIVINGAPARTMENTS_AVG', 'LIVINGAREA_AVG', 'NONLIVINGAPARTMENTS_AVG',
                        'NONLIVINGAREA_AVG', 'APARTMENTS_MODE', 'BASEMENTAREA_MODE', 'YEARS_BEGINEXPLUATATION_MODE',
                       'YEARS_BUILD_MODE', 'COMMONAREA_MODE', 'ELEVATORS_MODE', 'ENTRANCES_MODE', 'FLOORSMAX_MODE',
                        'FLOORSMIN_MODE', 'LANDAREA_MODE', 'LIVINGAPARTMENTS_MODE', 'LIVINGAREA_MODE',
                       'NONLIVINGAPARTMENTS_MODE', 'NONLIVINGAREA_MODE', 'APARTMENTS_MEDI', 'BASEMENTAREA_MEDI',
                        'YEARS_BEGINEXPLUATATION_MEDI', 'YEARS_BUILD_MEDI', 'COMMONAREA_MEDI', 'ELEVATORS_MEDI',
                        'ENTRANCES_MEDI', 'FLOORSMAX_MEDI', 'FLOORSMIN_MEDI', 'LANDAREA_MEDI', 'LIVINGAPARTMENTS_MEDI',
                        'LIVINGAREA_MEDI', 'NONLIVINGAPARTMENTS_MEDI', 'NONLIVINGAREA_MEDI', 'FONDKAPREMONT_MODE',
                        'HOUSETYPE_MODE', 'TOTALAREA_MODE', 'WALLSMATERIAL_MODE', 'EMERGENCYSTATE_MODE','DAYS_LAST_PHONE_CHANGE',
                       'ORGANIZATION_TYPE','FLAG_DOCUMENT_2','FLAG_DOCUMENT_3','FLAG_DOCUMENT_4', 'FLAG_DOCUMENT_5',
                        'FLAG_DOCUMENT_6','FLAG_DOCUMENT_7','FLAG_DOCUMENT_8','FLAG_DOCUMENT_9','FLAG_DOCUMENT_10',
                        'FLAG_DOCUMENT_11','FLAG_DOCUMENT_12','FLAG_DOCUMENT_13','FLAG_DOCUMENT_14','FLAG_DOCUMENT_15',
                        'FLAG_DOCUMENT_16','FLAG_DOCUMENT_17', 'FLAG_DOCUMENT_18','FLAG_DOCUMENT_19','FLAG_DOCUMENT_20',
                        'FLAG_DOCUMENT_21'
                       
                       ]), axis=1)

risk_1.info()
```

```python
# Paso 5: Analizar los valores nulos en cada variable para determinar qué decisiones tomar.

risk_1.isnull().sum()
```

```python
# Paso 6: Reemplazar los valores nulos por media o moda en función del tipo de variable (numérica o categórica).

risk_1['AMT_ANNUITY'] = risk_1['AMT_ANNUITY'].fillna(risk_1['AMT_ANNUITY'].mean())
risk_1['AMT_GOODS_PRICE'] = risk_1['AMT_GOODS_PRICE'].fillna(risk_1['AMT_GOODS_PRICE'].mean())
risk_1['CNT_FAM_MEMBERS'] = risk_1['CNT_FAM_MEMBERS'].fillna(risk_1['CNT_FAM_MEMBERS'].mean())
risk_1['OCCUPATION_TYPE'] = risk_1['OCCUPATION_TYPE'].fillna(risk_1['OCCUPATION_TYPE'].mode().iloc[0])
risk_1['EXT_SOURCE_1'] = risk_1['EXT_SOURCE_1'].fillna(risk_1['EXT_SOURCE_1'].mean())
risk_1['EXT_SOURCE_2'] = risk_1['EXT_SOURCE_2'].fillna(risk_1['EXT_SOURCE_2'].mean())
risk_1['EXT_SOURCE_3'] = risk_1['EXT_SOURCE_3'].fillna(risk_1['EXT_SOURCE_3'].mean())
risk_1['OBS_30_CNT_SOCIAL_CIRCLE'] = risk_1['OBS_30_CNT_SOCIAL_CIRCLE'].fillna(risk_1['OBS_30_CNT_SOCIAL_CIRCLE'].mean())
risk_1['DEF_30_CNT_SOCIAL_CIRCLE'] = risk_1['DEF_30_CNT_SOCIAL_CIRCLE'].fillna(risk_1['DEF_30_CNT_SOCIAL_CIRCLE'].mean()) 
risk_1['OBS_60_CNT_SOCIAL_CIRCLE'] = risk_1['OBS_60_CNT_SOCIAL_CIRCLE'].fillna(risk_1['OBS_60_CNT_SOCIAL_CIRCLE'].mean())
risk_1['DEF_60_CNT_SOCIAL_CIRCLE'] = risk_1['DEF_60_CNT_SOCIAL_CIRCLE'].fillna(risk_1['DEF_60_CNT_SOCIAL_CIRCLE'].mean()) 
risk_1['AMT_REQ_CREDIT_BUREAU_HOUR'] = risk_1['AMT_REQ_CREDIT_BUREAU_HOUR'].fillna(risk_1['AMT_REQ_CREDIT_BUREAU_HOUR'].mean())
risk_1['AMT_REQ_CREDIT_BUREAU_DAY'] = risk_1['AMT_REQ_CREDIT_BUREAU_DAY'].fillna(risk_1['AMT_REQ_CREDIT_BUREAU_DAY'].mean())
risk_1['AMT_REQ_CREDIT_BUREAU_WEEK'] = risk_1['AMT_REQ_CREDIT_BUREAU_WEEK'].fillna(risk_1['AMT_REQ_CREDIT_BUREAU_WEEK'].mean())
risk_1['AMT_REQ_CREDIT_BUREAU_MON'] = risk_1['AMT_REQ_CREDIT_BUREAU_MON'].fillna(risk_1['AMT_REQ_CREDIT_BUREAU_MON'].mean())
risk_1['AMT_REQ_CREDIT_BUREAU_QRT'] = risk_1['AMT_REQ_CREDIT_BUREAU_QRT'].fillna(risk_1['AMT_REQ_CREDIT_BUREAU_QRT'].mean())
risk_1['AMT_REQ_CREDIT_BUREAU_YEAR'] = risk_1['AMT_REQ_CREDIT_BUREAU_YEAR'].fillna(risk_1['AMT_REQ_CREDIT_BUREAU_YEAR'].mean())
risk_1['OWN_CAR_AGE'] = risk_1['OWN_CAR_AGE'].fillna(risk_1['OWN_CAR_AGE'].mean())
```

```python
# Paso 7: Comprobar que los valores nulos se han remplazado por la media y la moda para cada una de las variables.

risk_1.isnull().sum()
```

```python
# Paso 8: Comprobar si existen valores duplicados en el Dataframe

risk_1.duplicated().any()
```

```python
# Paso 9: Aplicar criterios de consistencia a las columnas con metodo lower() 

risk_1.columns = risk_1.columns.str.lower()

risk_1.columns
```

```python
# Paso 10: Renombrar columnas para identificar de manera más ágil cada variable.

risk_1.rename(columns={'sk_id_curr': 'id'}, inplace=True)
risk_1.rename(columns={'name_contract_type': 'loan_type'}, inplace=True)
risk_1.rename(columns={'code_gender': 'gender'}, inplace=True)
risk_1.rename(columns={'flag_own_car': 'own_car'}, inplace=True)
risk_1.rename(columns={'flag_own_realty': 'own_realty'}, inplace=True)
risk_1.rename(columns={'cnt_children': 'children'}, inplace=True)
risk_1.rename(columns={'amt_income_total': 'client_income'}, inplace=True)
risk_1.rename(columns={'amt_credit': 'credit_amount'}, inplace=True)
risk_1.rename(columns={'amt_annuity': 'loan_annuity'}, inplace=True)
risk_1.rename(columns={'amt_annuity': 'loan_annuity'}, inplace=True)
risk_1.rename(columns={'amt_goods_price': 'goods_price'}, inplace=True)
risk_1.rename(columns={'name_type_suite': 'type_suite'}, inplace=True)
risk_1.rename(columns={'name_income_type': 'client_income_type'}, inplace=True)
risk_1.rename(columns={'name_education_type': 'education'}, inplace=True)
risk_1.rename(columns={'name_family_status': 'family_status'}, inplace=True)
risk_1.rename(columns={'days_birth': 'days_birth'}, inplace=True)
risk_1.rename(columns={'name_housing_type': 'housing_situation'}, inplace=True)
risk_1.rename(columns={'own_car_age': 'car_age'}, inplace=True)
risk_1.rename(columns={'occupation_type': 'client_ocupation'}, inplace=True)
risk_1.rename(columns={'cnt_fam_members': 'family_members'}, inplace=True)
risk_1.rename(columns={'region_rating_client': 'region_rating'}, inplace=True)
risk_1.rename(columns={'region_rating_client_w_city': 'city_rating'}, inplace=True)
risk_1.rename(columns={'ext_source_1': 'ext_score_1'}, inplace=True)
risk_1.rename(columns={'ext_source_2': 'ext_score_2'}, inplace=True)
risk_1.rename(columns={'ext_source_3': 'ext_score_3'}, inplace=True)
risk_1.rename(columns={'obs_30_cnt_social_circle': 'obs_30'}, inplace=True)
risk_1.rename(columns={'def_30_cnt_social_circle': 'def_30'}, inplace=True)
risk_1.rename(columns={'obs_60_cnt_social_circle': 'obs_60'}, inplace=True)
risk_1.rename(columns={'def_60_cnt_social_circle': 'def_60'}, inplace=True)
risk_1.rename(columns={'amt_req_credit_bureau_hour': 'req_hour'}, inplace=True)
risk_1.rename(columns={'amt_req_credit_bureau_day': 'req_day'}, inplace=True)
risk_1.rename(columns={'amt_req_credit_bureau_week': 'req_week'}, inplace=True)
risk_1.rename(columns={'amt_req_credit_bureau_mon': 'req_mon'}, inplace=True)
risk_1.rename(columns={'amt_req_credit_bureau_qrt': 'req_qrt'}, inplace=True)
risk_1.rename(columns={'amt_req_credit_bureau_year': 'req_year'}, inplace=True)
```

```python
# Paso 11: Mejorar las consistencia interna de los valores únicos de cada variable.

# cambio en variable 'target': '1' por 'High Risk' y '0' por 'Low Risk'.

risk_1.target = risk_1.target.astype(str)
risk_1.target = risk_1.target.str.replace('1','High Risk').replace('0','Low Risk')

# cambio en variable 'own_car': 'N' por 'No' y 'Y' por 'Yes'.

risk_1.own_car = risk_1.own_car.str.replace('N','No').replace('Y','Yes')

# cambio en variable 'own_realty': 'N' por 'No' y 'Y' por 'Yes'.

risk_1.own_realty = risk_1.own_realty.str.replace('N','No').replace('Y','Yes')

# cambio en variable 'education': agrupación de algunas categorías ('Secondary').

risk_1.education = risk_1.education.str.replace('Secondary / secondary special','Secondary')

# cambio en variable 'family_status': agrupación de algunas categorías.

risk_1.family_status = risk_1.family_status.str.replace('Single / not married','Single').replace('Civil marriage', 'Married')

# cambio en variable 'housing_situation': agrupación de algunas categorías.

risk_1.housing_situation = risk_1.housing_situation.str.replace('House / apartment','Ownership').replace('With parents', 'Living with parents').replace('Co-op apartment','Cooperative apartment')

# cambio en variable 'client_ocupation': agrupación de algunas categorías.

risk_1.client_ocupation = risk_1.client_ocupation.str.replace('Low-skill Laborers','Laborers').replace('High skill tech staff', 'High Tech staff').replace('IT staff','High Tech staff')
risk_1.client_ocupation =  risk_1.client_ocupation.str.replace('staff','')

# cambio en variable 'client_income_type': agrupación de algunas categorías.

risk_1.client_income_type =  risk_1.client_income_type.str.replace('Commercial associate','Working').replace('Businessman','Working').replace('Maternity leave','Working')

# cambio en variable 'family_member': a int.

risk_1.family_members = risk_1.family_members.astype(int)
```

```python
# Paso 12: Analizar el dtype de cada variable para chequear si tiene sentido la info.

risk_1.info()
```

```python
# Paso 13: Redondear los valores de estas variables numéricas a 2 posiciones.

risk_1.car_age = risk_1.car_age.round(2) 
risk_1.ext_score_1 = risk_1.ext_score_1.round(2)
risk_1.ext_score_2 = risk_1.ext_score_2.round(2)
risk_1.ext_score_3 = risk_1.ext_score_3.round(2)
risk_1.req_hour = risk_1.req_hour.round(2)
risk_1.req_day = risk_1.req_day.round(2)
risk_1.req_week = risk_1.req_week.round(2)
risk_1.req_mon = risk_1.req_mon.round(2)
risk_1.req_qrt = risk_1.req_qrt.round(2)
risk_1.req_year = risk_1.req_year.round(2)
```

```python
# Paso 14: Transformar las variable de observaciones (defaulters y observables) en valores int.

risk_1.obs_30 = risk_1.obs_30.astype(int)
risk_1.def_30 = risk_1.def_30.astype(int)
risk_1.obs_60 = risk_1.obs_60.astype(int)
risk_1.def_60 = risk_1.def_60.astype(int)
```

```python
# Paso 15: Ajustar ineficiencias en las variables asociadas a las consultas del Área de Riesgo.

# variable 'req_hour': 

risk_1.req_hour = risk_1.req_hour.astype(str)
risk_1.req_hour = risk_1.req_hour.str.replace('0.01','0.0')
risk_1.req_hour = pd.to_numeric(risk_1.req_hour).astype('int32')

# variable 'req_day': 

risk_1.req_day = risk_1.req_day.astype(str)
risk_1.req_day = risk_1.req_day.str.replace('0.01','0.0')
risk_1.req_day = pd.to_numeric(risk_1.req_day).astype('int32')

# variable 'req_week':

risk_1.req_week = risk_1.req_week.astype(str)
risk_1.req_week = risk_1.req_week.str.replace('0.03','0.0')
risk_1.req_week = pd.to_numeric(risk_1.req_week).astype('int32')

# variable 'req_mon':

risk_1.req_mon = risk_1.req_mon.astype(str)
risk_1.req_mon = risk_1.req_mon.str.replace('0.27','0.0')
risk_1.req_mon = pd.to_numeric(risk_1.req_mon).astype('int32')


# variable 'req_qrt':

risk_1.req_qrt = risk_1.req_qrt.astype(str)
risk_1.req_qrt = risk_1.req_qrt.str.replace('0.01','0.0')
risk_1.req_qrt = pd.to_numeric(risk_1.req_qrt).astype('int32')

# variable 'req_year':

risk_1.req_year = risk_1.req_year.astype(str)
risk_1.req_year = risk_1.req_year.str.replace('1.9','2.0')
risk_1.req_year = pd.to_numeric(risk_1.req_year).astype('int32')
```

```python
# Paso 16: Eliminar incosistencias en valores desconocidos o no identificados:

# Eliminar valores XNA de variable 'gender'

risk_1 = risk_1[risk_1['gender'] != 'XNA']

# Eliminar valores Unknown de variable 'family_status'

risk_1 = risk_1[risk_1['family_status'] != 'Unknown']
```
