```python
# Paso 0: Importar las librerias necesarias para realizar el análisis del modelo

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report
import numpy as np
```

```python
# Paso 1: Crear un nuevo Dataframe del dataset limpio.

risk_tg = pd.read_csv('Model_2.csv')
```

```python
# Paso 2: Desplegar el Dataframe para chequear las variables iniciales

risk_tg
```

```python
# Paso 3: Testear el modelo con la nueva variable generada en el paso anterior

contingency_table = pd.crosstab(risk_tg['target'], risk_tg['category'], margins=True)

print(contingency_table)

print()

conf_matrix = confusion_matrix(risk_tg['target'], risk_tg['category'])

print(conf_matrix)

print()

# Classification report
print(classification_report(risk_tg['target'], risk_tg['category']))
```

```python
# Paso 4: Establecer una nueva clasificación modificando la probabilidad acumulada de default.

risk_tg['category'] = 'Low Risk'

# En vez de 0.6, utilizar un defaul de 0.65

conditions = (risk_tg['def_initial'] >= 0.65) | (risk_tg['def_5'] >= 0.65) | (risk_tg['def_10'] >= 0.65) | (risk_tg['def_15'] >= 0.65) | (risk_tg['def_20'] >= 0.65)
risk_tg.loc[conditions, 'category'] = 'High Risk'

risk_tg.category.value_counts()
```

```python
# Paso 5: Testear el modelo con la variable modificada.

contingency_table = pd.crosstab(risk_tg['target'], risk_tg['category'], margins=True)

print(contingency_table)

print()

conf_matrix = confusion_matrix(risk_tg['target'], risk_tg['category'])


print(conf_matrix)

print()

# Classification report
print(classification_report(risk_tg['target'], risk_tg['category']))
```
